```python
# LoopCurrent to place within 4 defined elements
with schemdraw.Drawing():
    R = elm.Resistor()
    C = elm.Capacitor().down()
    D = elm.Diode().left()
    L = elm.InductorIEC().up()
    elm.LoopCurrent([R, C, D, L], direction='ccw')
```

```python
# LoopArrow to place anywhere
with schemdraw.Drawing():
    R = elm.Potentiometer()
    elm.LoopArrow().at(R.start).color('red')
    elm.LoopArrow(direction='ccw').at(R.end).color('blue')
    elm.LoopArrow(width=2, height=1).at(R.tap).color('orange')
```
